```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# En supposant que 'df' est votre DataFrame contenant les donn√©es

# Question : Excluez les colonnes non num√©riques pour √©viter les erreurs dans le calcul de corr√©lation
numeric_df = df.select_dtypes(include=[np.number])  # s√©lectionne toutes les colonnes de type num√©rique (int, float)

# Question : Cr√©ez une figure avec une taille sp√©cifi√©e
plt.figure(figsize=(10, 6))

# Question : Calculez la matrice de corr√©lation uniquement pour les colonnes num√©riques
corr = numeric_df.corr()

# Question : Cr√©ez une carte de chaleur de la matrice de corr√©lation
sns.heatmap(corr, linewidths= 0.5, annot= True, fmt=".2f")   # linewidths : √©paisseur des lignes s√©parant les cases
# annot=True ‚Üí affiche les valeurs num√©riques dans chaque case
# fmt=".2f" ‚Üí format des nombres (2 d√©cimales)
plt.show()

# Multicolin√©arit√© : situation o√π deux ou plusieurs variables ind√©pendantes sont fortement corr√©l√©es.
# Matrice de corr√©lation : outil rapide pour d√©tecter la multicolin√©arit√©.

# Si toutes les corr√©lations sont faibles ou mod√©r√©es, il n‚Äôy a pas de probl√®me de multicolin√©arit√© √©vident.
# Si certaines corr√©lations sont tr√®s proches de 1 ou -1, il peut y avoir multicolin√©arit√©.

# Important : la corr√©lation n‚Äôest qu‚Äôun indicateur. Pour un diagnostic plus pr√©cis, on peut calculer le Variance Inflation Factor (VIF) pour chaque variable.
```

<p style="font-size: 20px; color: #34656d; font-family: 'Comic Sans MS';"><strong>Nous pouvons dire qu'il n'y a pas de multicolin√©arit√© pr√©sente dans les donn√©es.</strong></p>


```python
# Pour v√©rifier la multicolin√©arit√© de fa√ßon plus rigoureuse, on peut calculer le VIF (Variance Inflation Factor) pour chaque variable num√©rique.

import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# S√©lectionner uniquement les colonnes num√©riques
numeric_df = df.select_dtypes(include=[np.number])

# Cr√©er un DataFrame pour stocker les VIF
vif_data = pd.DataFrame()
vif_data["Variable"] = numeric_df.columns

# Calculer le VIF pour chaque colonne
vif_data["VIF"] = [variance_inflation_factor(numeric_df.values, i) 
                   for i in range(numeric_df.shape[1])]

print(vif_data)
```

// VIF > 5 (ou 10 selon la source) ‚Üí multicolin√©arit√© √©lev√©e ‚Üí certaines variables peuvent √™tre redondantes

```python
# Question : Importez la fonction to_categorical de tensorflow.keras.utils.
from tensorflow.keras.utils import to_categorical
# to_categorical : transforme des entiers (0, 1) en vecteurs binaires, par exemple : 0 ‚Üí [1, 0] // 1 ‚Üí [0, 1]

# Question : Cr√©ez le jeu de caract√©ristiques 'X' en supprimant la colonne 'Exited' du DataFrame 'df'.
X = df.drop('Exited', axis = 1)   # supprime la colonne cible Exited pour ne garder que les features

# Question : Cr√©ez le jeu d'√©tiquettes 'y' en convertissant la colonne 'Exited' en cat√©gories binaires.
y = to_categorical(df['Exited'])   # transforme la colonne cible en format compatible pour les mod√®les Keras (classification binaire).
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # 0 = tout, 1 = info, 2 = warning, 3 = erreurs
import tensorflow as tf
```

```python
# Question : Importez la fonction train_test_split depuis le module sklearn.model_selection.
from sklearn.model_selection import train_test_split

# Question : Divisez les donn√©es en ensembles d'entra√Ænement et de test (25% de donn√©es pour le test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)  # random_state : pour obtenir toujours la m√™me r√©partition lors de multiples ex√©cutions 
```

```python
# Question : Importez la classe StandardScaler depuis le module sklearn.preprocessing.
from sklearn.preprocessing import StandardScaler

# Question : Cr√©ez une instance de StandardScaler nomm√©e 'sc'.
sc = StandardScaler()

# Question : Appliquez le scaler 'sc' pour ajuster et transformer les donn√©es de l'ensemble d'entra√Ænement 'X_train'.
X_train = sc.fit_transform(X_train)

# Question : Utilisez le m√™me scaler pour transformer les donn√©es de l'ensemble de test 'X_test'.
X_test = sc.transform(X_test)
```

### Construire notre ANN brique par brique

```python
# Question : Importez keras et les modules n√©cessaires pour construire un r√©seau de neurones.
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization

# Question : Initialisez un r√©seau de neurones artificiels s√©quentiel.
model = Sequential()

# Question : Ajoutez la premi√®re couche d'entr√©e et la premi√®re couche cach√©e avec 10 neurones, 
# utilisez l'initialiseur 'normal' pour les poids et 'relu' comme fonction d'activation.
model.add(Dense(10, kernel_initializer = 'normal', activation = 'relu', input_shape = (X_train.shape[1], )))  # input_shape doit correspondre au nombre de features (colonnes) du jeu X_train.
# Apr√®s preprocessing, X_train contient toutes les colonnes num√©riques encod√©es

# Question : Ajoutez une couche de normalisation par lots et une couche de Dropout avec un taux de 0.1.
model.add(Dropout(rate = 0.1))
model.add(BatchNormalization())   # BatchNormalization() normalise les activations
# On met souvent BatchNormalization apr√®s la couche Dense et avant l‚Äôactivation, c'est une couche de normalisation utilis√©e dans les r√©seaux de neurones pour am√©liorer l‚Äôentra√Ænement et la stabilit√© du mod√®le.

# Question : Ajoutez la troisi√®me couche cach√©e avec 7 neurones, utilisez l'initialiseur 'normal' et 'relu' comme activation.
model.add(Dense(7, kernel_initializer = 'normal', activation = 'relu'))

# Question : R√©p√©tez l'ajout de la couche de Dropout et de normalisation par lots.
model.add(Dropout(rate = 0.1))
model.add(BatchNormalization())

# Question : Ajoutez la couche de sortie avec 2 neurones et 'sigmoid' comme fonction d'activation.
model.add(Dense(2, kernel_initializer = 'normal', activation = 'sigmoid'))

# Question : Compilez le mod√®le en utilisant 'adam' comme optimiseur, 'binary_crossentropy' comme fonction de perte et suivez la m√©trique 'accuracy'.
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Question : Entra√Ænez le mod√®le sur l'ensemble d'entra√Ænement avec une validation crois√©e de 20%, utilisez √©galement les donn√©es de test pour la validation, et r√©p√©tez pour 50 √©poques.
model_history = model.fit(X_train, y_train, validation_split = 0.2, validation_data = (X_test, y_test), epochs = 50)
```
